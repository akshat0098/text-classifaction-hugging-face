# Lets do some preprocessing

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
```

```python
def hashtags(text):
  hash = re.findall(r"#(\w+)", text)
  return 
  
def remove_users(tweet):
    '''Takes a string and removes retweet and @user information'''
    tweet = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', tweet) 
    # remove tweeted at
    return tweet

def remove_links(tweet):
    '''Takes a string and removes web links from it'''
    tweet = re.sub(r'http\S+', '', tweet) # remove http links
    tweet = re.sub(r'bit.ly/\S+', '', tweet) # rempve bitly links
    tweet = tweet.strip('[link]') # remove [links]
    return tweet

def clean_html(text):  
  html = re.compile('<.*?>')#regex
  return html.sub(r'',text)    

def non_ascii(s):
  return "".join(i for i in s if ord(i)<128)

def lower(text):
  return text.lower()

def removeStopWords(str):
  #select english stopwords
  cachedStopWords = set(stopwords.words("english"))
  #add custom words
  cachedStopWords.update(('and','I','A','http','And','So','arnt','This','When','It','many','Many','so','cant','Yes','yes','No','no','These','these','mailto','regards','ayanna','like','email'))
  #remove stop words
  new_str = ' '.join([word for word in str.split() if word not in cachedStopWords]) 
  return new_str

def email_address(text):
  email = re.compile(r'[\w\.-]+@[\w\.-]+')
  return email.sub(r'',text)

def punct(text):
  token=RegexpTokenizer(r'\w+')#regex
  text = token.tokenize(text)
  text= " ".join(text)
  return text 
def remove_digits(text):
    pattern = r'[^a-zA-z.,!?/:;\"\'\s]' 
    return re.sub(pattern, '', text)

def remove_special_characters(text):
    # define the pattern to keep
    pat = r'[^a-zA-z0-9.,!?/:;\"\'\s]' 
    return re.sub(pat, '', text)

def email_address(text):
  """   remove email """
  email = re.compile(r'[\w\.-]+@[\w\.-]+')
  return email.sub(r'',text)


def remove_links(text):
    '''Takes a string and removes web links from it'''
    
    text = re.sub(r'http\S+', '', text) # remove http links
    text = re.sub(r'bit.ly/\S+', '', text) # rempve bitly links
    text = re.sub(r'www\S+','',text)
    text =text.strip('[link]') # remove [links]
    return text



df['hashtag'] = df.text.apply(func = hashtags)
df['new_text'] = df.text.apply(func = email_address)
df['new_text'] = df.text.apply(func = remove_links)


df['new_text'] = df.text.apply(func = remove_users)
df['new_text'] = df.text.apply(func = clean_html)
df['new_text'] = df.text.apply(func = non_ascii)
df['new_text'] = df.text.apply(func = lower)
df['new_text'] = df.new_text.apply(func = email_address)
df['new_text'] = df.new_text.apply(func = removeStopWords)
df['new_text'] = df.new_text.apply(func = clean_html)
df['new_text'] = df.new_text.apply(func = punct)
```
